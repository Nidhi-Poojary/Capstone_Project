```python
#Q5
import pandas as pd
from IPython.display import display  # For clean table display in notebooks

# Load the data (adjust the filenames if needed)
female_df = pd.read_csv("/content/drive/MyDrive/Capstone_Project/nhanes_adult_female_bmx_2020.csv")
male_df = pd.read_csv("/content/drive/MyDrive/Capstone_Project/nhanes_adult_female_bmx_2020.csv")

# Descriptive stats for females
female_stats = female_df['weight'].describe()
female_stats['skewness'] = female_df['weight'].skew()
female_stats['kurtosis'] = female_df['weight'].kurt()
female_stats['25th'] = female_df['weight'].quantile(0.25)
female_stats['50th'] = female_df['weight'].quantile(0.50)
female_stats['75th'] = female_df['weight'].quantile(0.75)

# Descriptive stats for males
male_stats = male_df['weight'].describe()
male_stats['skewness'] = male_df['weight'].skew()
male_stats['kurtosis'] = male_df['weight'].kurt()
male_stats['25th'] = male_df['weight'].quantile(0.25)
male_stats['50th'] = male_df['weight'].quantile(0.50)
male_stats['75th'] = male_df['weight'].quantile(0.75)

# Summary table
summary_df = pd.DataFrame({
    'Mean': [female_stats['mean'], male_stats['mean']],
    'Median': [female_stats['50th'], male_stats['50th']],
    'Standard Deviation': [female_stats['std'], male_stats['std']],
    'Min': [female_stats['min'], male_stats['min']],
    'Max': [female_stats['max'], male_stats['max']],
    '25th Percentile': [female_stats['25th'], male_stats['25th']],
    '50th Percentile': [female_stats['50th'], male_stats['50th']],
    '75th Percentile': [female_stats['75th'], male_stats['75th']],
    'Skewness': [female_stats['skewness'], male_stats['skewness']],
    'Kurtosis': [female_stats['kurtosis'], male_stats['kurtosis']]
}, index=['Female', 'Male'])

summary_df = summary_df.round(2)
summary_df.index.name = 'Gender'

# Display nicely formatted table
print(" Summary of Weight Distributions:\n")
display(summary_df)

# Interpretation text
print("\n Interpretation:")
print("- Both male and female weight distributions exhibit positive skewness, indicating a tail extending towards higher weights.")
print("- The male distribution has a slightly higher mean and median weight than the female distribution, suggesting that males tend to weigh more on average.")
print("- Both distributions have similar standard deviations, but male weights show a wider range (Min to Max).")
print("- The kurtosis values indicate that both distributions are slightly more peaked than a normal distribution (leptokurtic).")
```

```python
#Q8
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore, pearsonr, spearmanr

df['BMI'] = df['weight'] / ((df['std height'] / 100) ** 2)

cols = ['std height', 'weight', 'waist circum', 'hip- circum', 'BMI']
zfemale = df[cols].apply(zscore)
zfemale.columns = [f"{col}_zscore" for col in cols]  # Rename columns for clarity

def corrfunc(x, y, **kws):
    pearson_r, pearson_p = pearsonr(x, y)
    spearman_r, spearman_p = spearmanr(x, y)
    ax = plt.gca()
    ax.annotate(
        f'Pearson: {pearson_r:.2f} (p={pearson_p:.2f})\nSpearman: {spearman_r:.2f} (p={spearman_p:.2f})',
        xy=(.5, .5), xycoords=ax.transAxes,
        ha='center', va='center', fontsize=8,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.5)
    )

g = sns.PairGrid(zfemale)
g.map_lower(sns.scatterplot, s=20)
g.map_diag(sns.histplot, kde=True)
g.map_upper(corrfunc)

plt.suptitle("Pairplot of Standardized Female Data with Correlations", y=1.03)
plt.show()
```

```python
#Q10
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


male_df = pd.read_csv("/content/drive/MyDrive/Capstone_Project/nhanes_adult_male_bmx_2020.csv")
female_df = pd.read_csv("/content/drive/MyDrive/Capstone_Project/nhanes_adult_female_bmx_2020.csv")

male_df.columns = [col.strip().lower().replace(' ', '_').replace('-', '') for col in male_df.columns]
female_df.columns = [col.strip().lower().replace(' ', '_').replace('-', '') for col in female_df.columns]

male_df["waist_to_height"] = male_df["waist_circum"] / male_df["std_height"]
male_df["waist_to_hip"] = male_df["waist_circum"] / male_df["hip_circum"]
female_df["waist_to_height"] = female_df["waist_circum"] / female_df["std_height"]
female_df["waist_to_hip"] = female_df["waist_circum"] / female_df["hip_circum"]


plot_df = pd.DataFrame({
    "Waist-to-Height": pd.concat([male_df["waist_to_height"], female_df["waist_to_height"]], ignore_index=True),
    "Waist-to-Hip": pd.concat([male_df["waist_to_hip"], female_df["waist_to_hip"]], ignore_index=True),
    "Gender": ["Male"] * len(male_df) + ["Female"] * len(female_df)
})


plot_df_melted = pd.melt(plot_df, id_vars="Gender", var_name="Ratio Type", value_name="Ratio")


plt.figure(figsize=(10, 6))
sns.boxplot(data=plot_df_melted, x="Ratio Type", y="Ratio", hue="Gender", palette="Set2")
plt.title("Distribution of Waist-to-Height and Waist-to-Hip Ratios by Gender")
plt.xlabel("Ratio Type")
plt.ylabel("Ratio")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Q11
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Metric": ["BMI", "Waist-to-Height Ratio (WHtR)", "Waist-to-Hip Ratio (WHR)"],
    "Advantages": [
        " Easy to calculate\n Widely used\n Good for screening",
        " Reflects fat distribution\n Predicts health risks better\n One universal cut-off",
        " Shows central obesity\n Related to heart risk\n Simple if measured correctly"
    ],
    "Disadvantages": [
        " Doesn't show fat location\n Misleading for muscular people\n Less accurate for some",
        " Needs waist & height\n Not widely used clinically",
        " Needs accurate waist & hip\n Cut-offs vary by sex\n Can be misleading for some"
    ]
}


df = pd.DataFrame(data)

print("Advantages and Disadvantages Table:")
display(df.style.set_properties(**{'white-space': 'pre-wrap'}).set_caption("Comparison of BMI, WHtR, and WHR"))


df['Pro Count'] = df['Advantages'].apply(lambda x: x.count(''))
df['Con Count'] = df['Disadvantages'].apply(lambda x: x.count(''))


plt.figure(figsize=(8, 5))
plt.bar(df["Metric"], df["Pro Count"], label="Advantages", color='green', alpha=0.7)
plt.bar(df["Metric"], df["Con Count"], label="Disadvantages", color='red', alpha=0.7, bottom=df["Pro Count"])
plt.title("Number of Advantages and Disadvantages per Metric")
plt.ylabel("Count")
plt.legend()
plt.xticks(rotation=15)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
#Q12
import numpy as np
import pandas as pd
from scipy.stats import zscore
from google.colab import files
from IPython.display import display


try:
    df = pd.read_csv('/content/drive/MyDrive/Capstone_Project/nhanes_adult_female_bmx_2020.csv')
except FileNotFoundError:
    print("File not found.")
    exit()
except Exception as e:
    print(f"Error: {e}")
    exit()


df['BMI'] = df['weight'] / ((df['std height'] / 100) ** 2)


cols_to_standardize = ['std height', 'weight', 'waist circum', 'hip- circum', 'BMI']
zfemale = df[cols_to_standardize].apply(zscore)

sorted_bmi_indices = np.argsort(zfemale['BMI'])
lowest_bmi_indices = sorted_bmi_indices[:5]
highest_bmi_indices = sorted_bmi_indices[-5:]

lowest_z = zfemale.loc[lowest_bmi_indices].copy()
lowest_z["BMI Group"] = "Lowest"

highest_z = zfemale.loc[highest_bmi_indices].copy()
highest_z["BMI Group"] = "Highest"

combined_z = pd.concat([lowest_z, highest_z], axis=0).reset_index(drop=True)

print("Standardized Body Measurements for Lowest and Highest BMI Individuals:")
display(combined_z)


print("\nInterpretation:")
print("- Lowest BMI group: All values are mostly negative (below average).")
print("- Highest BMI group: Most values are strongly positive (above average).")
print("- Height varies slightly in both groups; BMI mostly driven by weight and circumferences.")
```
